```python
iris=sn.load_dataset('iris')
from matplotlib import pyplot as plt
x=iris.species
y=iris.petal_length
plt.bar(x,y)
plt.show()
```

```python
y=iris.species
x1=iris.sepal_length
x2=iris.sepal_width
fig=plt.figure()
ax=fig.add_axes([0,0,2,1])
l1=ax.plot(x1,y,'y-')
l2=ax.plot(x2,y,'g--')
ax.legend(labels=('Line1','Line2'),loc='upper right')
ax.set_title("add axes")
ax.set_xlabel("xaxis")
ax.set_ylabel("yaxis")
plt.show()
```
